```python
#Import libraries:
import librosa
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
import IPython.display as ipd
```

```python
#load .wav file:
audio_path='./genres/rock/rock.00001.wav'
```

```python
#This returns an audio time series as a numpy array with a default sampling rate(sr) of 22KHZ mono:
x , sr = librosa.load(audio_path)
```

```python
ipd.Audio(audio_path)
```

```python
#This gives us a numpy array of approx. 661500 elements. This represents each audio clip, but it would be 
#too much to handle. We would end up with a 1000x661500 dataframe. That's more than half a million features
#to predict one label (overdimensioned), perhaps too much for any ML model to handle.
print(x.shape, sr)
```

```python
#Plot the amplitude envelope of the waveform:
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```
